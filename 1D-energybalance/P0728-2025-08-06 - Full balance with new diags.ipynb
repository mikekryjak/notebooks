- Neutral viscosity should take CX and IZ 
- Ion viscosity should take ii

```python
fig, axes = plt.subplots(4,1, figsize = (10, 2.5*4), sharex = True)
fig.subplots_adjust(hspace=0)
species = "e"

ds = cs["base"].ds.isel(t=-1)
dom = ds.isel(pos = slice(2,-1))
x = dom["pos"].values.max() - dom["pos"].values + 1e-3
style = dict(marker = "+", ms = 2, lw = 0.75)
titlepos = dict(x = 0.5, y = 0.86)

### Total
ax = axes[0]
ax.set_title("Total", **titlepos)
electrons = dom[f"efe_tot_ylow"]
ions = dom[f"efd+_tot_ylow"]
neutrals = dom[f"efd_tot_ylow"]
radiation = (abs(dom["Rd+_ex"]) + abs(dom["Rd+_rec"]))*dom["dv"]
radiation[-1] = 0 # Domain has one guard cell where volumetric quantities are zero
source = 3/2 * (dom["Pd+_src"] + dom["Pe_src"]) * dom["dv"]
total = electrons + ions + neutrals + radiation + source

ax.plot(x, electrons,  c = "b", label = "Electrons", **style)
ax.plot(x, ions,  c = "r", label = "Ions", **style)
ax.plot(x, neutrals,  c = "darkslategrey", label = "Neutrals", **style)
ax.plot(x, source, label = "Source", **style)
ax.plot(x, radiation,  c = "deeppink", label = "Radiation \nintegral\n(IZ+REC)", **style)
ax.plot(x, total,  c = "black", label = "Total", **style)

### Electrons
ax = axes[1]
ax.set_title("Electrons", **titlepos)

ax.plot(x, dom[f"efe_cond_ylow"], **style, label = "Conduction")
ax.plot(x, dom[f"efe_adv_ylow"], **style, label = "Advection")
ax.plot(x, dom[f"efe_tot_ylow"], **style, c ="black", label = "Total")

### Ions
ax = axes[2]
ax.set_title("Ions", **titlepos)
conduction = dom[f"efd+_cond_ylow"]
advection = dom[f"efd+_adv_ylow"]
kinetic = 0.5 * 2*constants("mass_p") * dom[f"Vd+"] **2 
Qei = dom["Kd+e_coll"] * dom["dv"]
friction_e = dom["Ed+d_coll_friction"] * dom["dv"]
CX = -dom["Edd+_cx"] * dom["dv"]

ax.plot(x, conduction, **style, label = "Conduction")
ax.plot(x, advection, **style, label = "Advection")
ax.plot(x, kinetic, **style, label = "Kinetic energy")
ax.plot(x, Qei, **style, label = "Qei")
ax.plot(x, CX, **style, label = "CX")
ax.plot(x, dom[f"efd+_tot_ylow"], c = "black", **style, label = "Total")

### Neutrals
ax = axes[3]
ax.set_title("Neutrals", **titlepos)
kinetic = 0.5 * 2*constants("mass_p") * dom[f"Vd"] **2 

ax.plot(x, dom[f"efd_cond_ylow"], **style, label = "Conduction")
ax.plot(x, dom[f"efd_adv_ylow"], **style, label = "Advection")
ax.plot(x, kinetic, **style, label = "Kinetic")
ax.plot(x, dom[f"efd_tot_ylow"], c = "black", **style, label = "Total")

for ax in axes:
    ax.legend(bbox_to_anchor = (1, 1), loc = "upper left")
    ax.set_xscale("log")
    ax.set_ylabel("Power [W]")
    ax.grid()


```
