```python
casestore = dict()
```

```python
db = CaseDB(
    case_dir = r"/ssd_scratch/hcases",
    grid_dir = r"/ssd_scratch/hcases"
    )

toload = [
    dict(name="base", id="rmult1a-lores_base", unnormalise_geom = True),  
    dict(name="new", id="rmult2a-scale_iz", unnormalise_geom = True), 
]

for case in toload:
    casestore[case["name"]] = Load.case_1D(db.casepaths[case["id"]])
```

# Charge exchange

```python
fig, ax = plt.subplots(dpi = 100)

base = casestore["base"].ds.isel(pos=slice(2,-2))
new = casestore["new"].ds.isel(pos=slice(2,-2))

base["Fdd+_cx"].plot(ax = ax, label = "base")
(new["Fdd+_cx"]/0.9).plot(ax = ax, label = "new / multiplier")
ax.legend()

ax.set_xlim(11.5, 12.0)
```
