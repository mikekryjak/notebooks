# Scaling wall length

```python
side = "iu"
design = "SPR38"


d_original= eqb[side][design]
d = copy.deepcopy(d_original)

Xpoint = d["Xpoint"]
Sx = d["S"][Xpoint]
Bx = d["Btot"][Xpoint]

len_fac = 0.5

# find index to cut up to
target_length = Sx * (1-len_fac)
cut_idx = np.argmin(np.abs(d["S"] - target_length))

for param in ["S", "Spol", "Btot", "Bpol"]:
    d[param] = d[param][cut_idx:]
    
for param in ["S", "Spol"]:
    d[param] -= d[param][0]

d["Xpoint"] = d["Xpoint"] - cut_idx

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(d_original["S"], d_original["Btot"], linewidth = 3, label = "original", color = "black")
ax.plot(d["S"], d["Btot"], linewidth = 3, label = "cut", linestyle = "dashed", color = "red")

ax.scatter(d_original["S"][d_original["Xpoint"]], d_original["Btot"][d_original["Xpoint"]], color = "black", marker = "x", s= 200,linewidth =3,  label = "Original X-point", zorder = 10)
ax.scatter(d["S"][d["Xpoint"]], d["Btot"][d["Xpoint"]], color = "red", marker = "+", s= 300,linewidth =3,  label = "Cut X-point", zorder = 10)

ax.scatter(d_original["S"][0], d_original["Btot"][0], color = "black", marker = "o", s= 50,linewidth =3, zorder = 10, label = "Original wall")
ax.scatter(d["S"][0], d["Btot"][0], color = "red", marker = "o", s= 50,linewidth =3, zorder = 10, label = "Cut wall")

ax.set_title(side, fontsize = 15)
ax.legend(fontsize = 11)
ax.set_xlabel("S: parallel distance"); 
ax.set_ylabel("Btot")
ax.grid(which="both")
```

```python
side = "iu"
design = "SPR38"

def cut_wall(d_input, len_fac):

    d = copy.deepcopy(d_input)

    Xpoint = d["Xpoint"]
    Sx = d["S"][Xpoint]
    Bx = d["Btot"][Xpoint]

    # find index to cut up to
    target_length = Sx * (1-len_fac)
    cut_idx = np.argmin(np.abs(d["S"] - target_length))

    for param in ["S", "Spol", "Btot", "Bpol"]:
        d[param] = d[param][cut_idx:]
        
    for param in ["S", "Spol"]:
        d[param] -= d[param][0]

    d["Xpoint"] = d["Xpoint"] - cut_idx
    
    return d
```

```python
# list_cuts = np.arange(0.1,1.1, 0.1)
list_cuts = [0.1, 0.5, 0.9, 1.0]

data_cuts = dict()
design = "SPR38"
stats = dict()

for side in ["iu", "ou"]:
    data_cuts[side] = dict()
    stats[side] = defaultdict(list)
    
    for cut in list_cuts:
        data_cuts[side][cut] = cut_wall(eqb[side][design], cut)
        d = data_cuts[side][cut]
        
        stats[side]["Xpoint"].append(d["Xpoint"])
        stats[side]["Lx"].append(d["S"][d["Xpoint"]])
        stats[side]["Bx"].append(d["Btot"][d["Xpoint"]])
        stats[side]["Bt"].append(d["Btot"][0])
        stats[side]["BxBt"].append(d["Btot"][d["Xpoint"]] / d["Btot"][0])
        stats[side]["len_fac"].append(cut) # length factor
```
